```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
import numpy as np
import os
```

```python
for folder in os.listdir('images'): #images ki jagag folder directory jha pe image download h
    img_dir =  os.path.join('images',folder)
    for img in os.listdir(img_dir):
       
        image = load_img(os.path.join(r'C:\Users\thaku\Downloads',img_dir,img)) 
        
        image = img_to_array(image)
        
        image = np.expand_dims(image, axis=0)


        aug = ImageDataGenerator(
                                rotation_range=45,
                                zoom_range=0.20,
                                fill_mode="nearest")
        total = 0

        print("[INFO] generating images...")
        imageGen = aug.flow(image, batch_size=1, save_to_dir=os.path.join(r'C:\Users\thaku\Downloads',img_dir),
                            save_prefix="image", save_format="jpg")
        
        for image in imageGen:
            total+=1
    
            if total==2:
                break
```
